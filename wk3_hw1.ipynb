```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('co2_weekly_mlo.txt', skiprows=49,
                     names=['yr', 'mon', 'day', 'decimal', 'ppm', ' #days', '1 yr ago', '10 yr ago', 'since 1800'],
                     delim_whitespace=True)
clean_df = df[df.ppm != -999.99]
clean_df.head()
pp = np.polynomial.Polynomial(np.polyfit(clean_df.decimal, clean_df.ppm, 1))
print(pp)
slope, intercept = pp
print(intercept, "", slope)
plt.scatter(clean_df.decimal, clean_df.ppm)
plt.show()
plt.plot(clean_df.decimal, clean_df.decimal*slope + intercept, color='red', label= 'Fit')
plt.show()
plt.plot(clean_df.decimal, pp(clean_df.decimal), color='red')
plt.show()
```

```python
### Problem 4

#- [ ] For the Seaborn example
#   - [ ] Change the order to: 2 or 3


import seaborn as sns; sns.set_theme(color_codes=True)
import pandas as pd
import matplotlib.pyplot as plt

# Get Data
df = pd.read_csv('co2_weekly_mlo.txt', skiprows=49,
                 names=['yr', 'mon', 'day', 'decimal','ppm', '#days', '1 yr ago', '10 yr ago', 'since 1800'],
                 delim_whitespace=True)
clean_df = df[df.ppm != -999.99]
# Set the plots
ax = sns.regplot(x='decimal', y='ppm', data=clean_df, ci=95, order=2,
                 line_kws={'label': '2nd order Linear Regression Line', 'color': 'm'},
                 label="CO2 Weekly average")
ax.set_ylabel("CO2 (ppm)")
ax.set_xlabel("Year")
ax.legend(loc="upper left")
plt.show()
```

```python
### Problem 5 (a)
#   - [ ] Add x_bins = 100
#   - [ ] Add x_bins = 50

ax = sns.regplot(x='decimal', y='ppm', data=clean_df, ci=95, order=1, x_bins=100,
                 line_kws={'label': 'Linear regression line: $Y(X)=-3827 + 1.8 X$', 'color': 'm'},
                 label="CO2 Weekly average")
ax.set_ylabel("CO2 (ppm)")
ax.set_xlabel("Year")
ax.legend(loc="upper left")
plt.show()
```

```python
### Problem 5 (b)

#   - [ ] Add x_bins = 50

ax = sns.regplot(x='decimal', y='ppm', data=clean_df, ci=95, order=2, x_bins=50,
                 line_kws={'label': 'Linear regression line: $Y(X)=-3827 + 1.8 X$', 'color': 'm'},
                 label="CO2 Weekly average")
ax.set_ylabel("CO2 (ppm)")
ax.set_xlabel("Year")
ax.legend(loc="upper left")
plt.show()
```
